## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

Check skewness and apply transformations

Convert data types if needed

Feature transformations (date parts, derived fields if required for analysis)

```python
# Handling missing values

df.isnull().sum()
```

```python
# Handling Numerical Missing Values - Discount and Shipping cost - Using 'Mean' as an imputation values since it provides an summarized value

df['discount'].fillna(df['discount'].mean(), inplace=True)
df['shipping_cost'].fillna(df['shipping_cost'].mean(), inplace=True)

df.isnull().sum()
```

```python
#Handling Categorical Missing Values - Payment method, Region and Gender - Using 'Mode' for imputation beacuse the value obtained is most frequent

df['payment_method'].fillna(df['payment_method'].mode()[0], inplace=True)
df['region'].fillna(df['region'].mode()[0], inplace=True)
df['customer_gender'].fillna(df['customer_gender'].mode()[0], inplace=True)

df.isnull().sum()
```

**Note** - ***My dataset is free of duplicate values, ensuring the accuracy and reliability of the data for our analysis.***

```python
#Treating Outliers

df.describe().round(2)
```

```python
df.head()
```

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns   #getting only numerical columns for finding outliers
numerical_columns
```

```python
#identifying outliers using Boxplots

plt.figure(figsize=(13,10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)   # 4 rows, 4 columns layout
    sns.boxplot(x=df[col])
    plt.title(f'Distribution of {col}')

plt.suptitle('Boxplots of Numerical Columns', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
#identifying outliers using KDEplots

plt.figure(figsize=(13,10))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)   # 4 rows, 4 columns layout
    sns.kdeplot(x=df[col])
    plt.title(f'Distribution of {col}')

plt.suptitle('KDE Plots of Numerical Columns', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
#Identifying which value to use to treat the outliers in price column

df['price'].quantile(0.95)
```

```python
#Identifying shape of outliers in Price column using boolean indexing & slicing

pr_threshold = df['price'].quantile(0.95)

df[df['price'] > pr_threshold].shape
```

```python
# Handling the outlier values using 95th quartile using clip()

df['price'] = df['price'].clip(upper=pr_threshold)
```

```python
# Since the price changed, changing the correct values in total_amount

df['total_amount'] = df['price'] * df['quantity'] * (1 - df['discount'])
```

```python
#Stats after handled outliers

df[['price','total_amount']].describe()
```

```python
# Boxplot after handling outliers

sns.boxplot(x=df['price'])
plt.show()
```

```python
# KDE plot after handling outliers

sns.kdeplot(x=df['price'])
plt.show()
```

**Note**: ***Some outliers still remain, but this is expected in an e-commerce dataset and does not pose a concern.***

```python
# Feature Transformation

# Date type conversion

df['order_date'] = pd.to_datetime(df['order_date'])
df['order_date']
```

```python
# Adding 3 more columns - Year, Month and day for analysis purposes

df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['day'] = df['order_date'].dt.day

df.head()
```

##Univariate Analysis

```python
# Chart 1: Price Distribution using Histogram Plot
plt.figure(figsize=(10,6))

sns.histplot(
    data=df,
    x='price',
    bins=30,
    kde=True,
    hue='category',        # adds grouping
    multiple='stack',      # stacks distributions
    palette='Set2',        # color style
    stat='count'           # shows frequency
)

plt.title("Distribution of Product Prices by Category")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A histogram with KDE is used to visualize the distribution of product prices. The hue parameter is used to differentiate price distributions across product categories.

**Key Observations**:

The distribution remains right-skewed, with most products priced lower, while some categories contribute to higher price ranges.

**Features Used**:

The price column is used as the main variable, and category is used as a grouping feature to compare how pricing varies across different product categories.

**Business Insight**:

Different product categories have varying price ranges, but overall, the majority of products fall into the lower price segment, indicating a focus on affordable items.


```python
# Chart 2: Category Distribution using Count Plot

plt.figure(figsize=(10,6))

ax = sns.countplot(
    data=df,
    x='category',
    palette='Set2',
    order=df['category'].value_counts().index
)

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container)

plt.title("Distribution of Product Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A countplot is used to visualize the frequency of each product category in the dataset. It helps understand how the data is distributed across different categories.

**Key Observations**:

Some categories have significantly higher counts compared to others, indicating that certain product types are more frequently purchased or listed in the dataset.

**Features Used**:

The category column is used as it represents the classification of products, and it is important for analyzing product popularity and sales distribution.

**Business Insight**:

The business appears to focus more on specific product categories, which may indicate higher demand or better sales performance in those segments.

```python
# Chart 3: Payment Method Distribution using Pie Plot

plt.figure(figsize=(7,7))

payment_counts = df['payment_method'].value_counts()

plt.pie(
    payment_counts,
    labels=payment_counts.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('Set2'),
    startangle=90,
    explode=[0.1,0.1,0,0,0,0],
    shadow=True
)

plt.title("Distribution of Payment Methods")
plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A pie chart is used to represent the proportion of different payment methods, with the most frequently used method highlighted using the explode feature.

**Key Observations**:

The highlighted segment clearly shows the most preferred payment method, while the remaining methods contribute smaller shares.

**Features Used**:

The payment_method column is used to analyze customer transaction preferences, and the explode feature is applied to emphasize the dominant category.

**Business Insight**:

The business can prioritize and optimize the most used payment method to improve customer satisfaction and ensure smoother transaction experiences.

##Bivariate Analysis

```python
# Chart 4: Impact of Product Price on Total Transaction Value using scatter plot

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x='price',
    y='total_amount',
    hue='category',
    palette='Set2',
    alpha=0.7,
    style='category',      # adds different markers
    markers=True           # enables marker variation
)

plt.title("Impact of Product Price on Total Transaction Value")
plt.xlabel("Price")
plt.ylabel("Total Amount")
plt.legend(bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A scatter plot is used to analyze the relationship between price and total_amount. It helps visualize how changes in product price impact the total transaction value. The hue parameter is used to differentiate product categories.

**Key Observations**:

There is a positive relationship between price and total amount, meaning as the price increases, the total transaction value also tends to increase. However, the spread indicates variability due to quantity and discount.

**Features Used**:

The price and total_amount columns are used as they directly represent product cost and total transaction value. The category feature is added to observe how this relationship varies across different product types.

**Business Insight**:

Higher-priced products contribute to larger transaction values, but variations suggest that factors like quantity and discounts also influence total sales. This helps the business understand revenue drivers.

```python
# Chart 5: Total Sales by Product Category using Bar plot

plt.figure(figsize=(12,6))

ax = sns.barplot(
    data=df,
    x='category',
    y='total_amount',
    estimator=sum,                 # total sales per category
    palette='Set2',
    errorbar=None,
    order=df.groupby('category')['total_amount'].sum().sort_values(ascending=False).index  # arranging categories based on total sales
)

# Adding bar labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', fontsize=9)

plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.xticks(rotation=45)
plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A bar plot is used to compare the total sales (total_amount) across different product categories. The estimator=sum is applied to aggregate total revenue for each category.

**Key Observations**:

Some categories generate significantly higher total sales compared to others, indicating uneven contribution across product segments.

**Features Used**:

The category column is used to group products, while total_amount is used as it represents the revenue generated from each transaction.

**Business Insight**:

The business can identify top-performing categories that drive the majority of revenue and focus on expanding or promoting those segments for increased profitability.

```python
# Chart 6: Sales Distribution Across Regions (Without Outliers) using Box Plot

plt.figure(figsize=(12,6))

sns.boxplot(
    data=df,
    x='region',
    y='total_amount',
    palette='Set2',
    showfliers=False   # removes outliers from display
)

plt.title("Sales Distribution Across Regions (Without Outliers)")
plt.xlabel("Region")
plt.ylabel("Total Amount")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Interpretation of above chart

**Chart Overview**:

A box plot is used to visualize the distribution of total sales across regions, with outliers removed to focus on the central data distribution.

**Key Observations**:

The spread of values across regions becomes clearer without extreme values, making it easier to compare median sales and variability.

**Features Used**:

The region column is used for comparison across geographical areas, while total_amount represents transaction values.

**Business Insight**:

By removing outliers from visualization, the business can better understand typical sales patterns across regions and make more consistent comparisons.

```python
# Chart 7A: Monthly Sales Trend

plt.figure(figsize=(12,6))

# Extract month
df['month'] = df['order_date'].dt.month

# Aggregate sales by month
monthly_sales = df.groupby('month')['total_amount'].sum()

# Plot
plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o'
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.grid(True)
plt.show()
```

```python
# Chart 7B: Yearly Sales Trend

plt.figure(figsize=(10,5))

# Extract year
df['year'] = df['order_date'].dt.year

# Aggregate sales by year
yearly_sales = df.groupby('year')['total_amount'].sum()

# Plot
plt.plot(
    yearly_sales.index,
    yearly_sales.values,
    marker='o'
)

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")

plt.grid(True)
plt.show()
```

## Interpretation of above charts (Monthly + Yearly Sales Trend)

**Chart Overview**:

Line charts are used to analyze sales trends over time at both monthly and yearly levels. Monthly analysis captures short-term variations, while yearly analysis provides an overall long-term trend.

**Key Observations**:

The monthly trend shows fluctuations in sales, indicating periods of higher and lower customer activity, which may reflect seasonal demand patterns. The yearly trend highlights the overall direction of sales performance, showing whether the business is growing, stable, or declining over time.

**Features Used**:

The order_date column is used to extract both month and year, while total_amount is used as it represents total sales aggregated over time.

**Business Insight**:

The combined analysis helps the business identify seasonal trends as well as long-term growth patterns. This enables better planning for inventory, marketing campaigns, and resource allocation during peak and low sales periods.

```python
# Chart 8: Correlation Heatmap of Numerical Features

plt.figure(figsize=(12,8))

# Select numerical columns
corr = df[numerical_columns].corr()

sns.heatmap(
    corr,
    annot=True,        # shows values
    cmap='coolwarm',   # color theme
    fmt='.2f',         # 2 decimal points
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features")
plt.xticks(rotation=45)
plt.show()
```

**Features Used**:

All numerical columns such as price, quantity, total_amount, discount, profit_margin, and others are used to analyze relationships across multiple variables simultaneously.

**Business Insight**:

**Key Observations**:

**Features Used**:
